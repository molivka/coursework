Для начала стоит сказать, что распределения случайной величины делятся на дискретные и непрерывные.


## Дискретное распределение

Случайная величина называется **дискретной**, если она принимает конечное или счётное количество значений.

**Дискретное распределение случайной величины** -- это функция, устанавливающая взаимосвязь между значениями, которые принимает случайная величина и соответствующими им вероятностями, причём количество принимаемых значений конечно или счётно в силу определения дискретной величины.

## Определения



**Функция распределения** -- это функция, которая характеризует распределение случайной величины. Она отражает вероятность того, что случайная величина примет значение, меньшее или равное определённому числу.

**Функция вероятности** -- это функция, которая описывает вероятность того, что дискретная случайная величина примет определённое значение.

**Функция плотности вероятности** -- это функция, которая описывает вероятность того, что непрерывная случайная величина примет определенное значение. То есть, функция плотности вероятности является аналогом функции вероятности, но для непрерывных величин.

## Генерация графиков распределений

###### Преамбула

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, geom, poisson, norm, uniform, expon, chi2

# кол-во генерируемых картинок
m = 10
```

### Дискретные

#### Биномиальное распределение

```python
for i in range(m):
  # кол-во испытаний
  n = np.random.randint(low=2, high=100)
  # кол-во успехов
  x = np.arange(n + 1)
  # вероятность успеха
  p = np.random.random(1)[0]
  # подсчёт вероятностной функции
  y = binom.pmf(x, n, p)
  # отрисовка
  plt.bar(x, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./binom/binom_{i}.jpg")
  plt.show()
  plt.close()
```

#### Геометрическое распределение

```python
for i in range(m):
  n = np.random.randint(low=1, high=100)
  # на каком испытании успех
  x = np.arange(n + 1)
  # вероятность одного успеха
  p = np.random.random(1)[0]
  # подсчёт вероятностной функции
  y = geom.pmf(x, p)
  # отрисовка
  plt.bar(x, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./geom/geom_{i}.jpg")
  plt.show()
  plt.close()
```

#### Распределение Пуассона

```python
for i in range(m):
  # число событий
  n = np.random.randint(low=1, high=100)
  # кол-во событий
  x = np.arange(n + 1)
  # параметр распределения
  lam = np.random.uniform(low=0, high=5)
  # подсчёт вероятностной функции
  y = poisson.pmf(x, lam)
  # отрисовка
  plt.bar(x, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./poisson/poisson_{i}.jpg")
  plt.show()
  plt.close()
```

### Непрерывные

#### Нормальное распределение

```python
for i in range(m):
  x1 = np.random.randint(1, 100)
  # среднее
  mu = np.random.randint(-x1 // 2, x1 // 2)
  # на каком отрезке сторим
  x = np.linspace(-x1, x1, 1000)
  # подсчёт вероятностной функции
  y = norm.pdf(x, loc=mu)
  # отрисовка
  plt.plot(x, y)
  plt.fill_between(x, 0, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./norm/norm_{i}.jpg")
  plt.show()
  plt.close()
```

#### Равномерное распределение

```python
for i in range(m):
  x1 = np.random.random(1)[0]
  # на каком отрезке сторим
  x = np.linspace(0, x1, 100)
  # подсчёт вероятностной функции
  y = uniform.pdf(x)
  # отрисовка
  plt.plot(x, y)
  plt.fill_between(x, 0, y)
  plt.xlim(-1, 1)
  # сохранение картинки в нужную папку
  plt.savefig(f"./uniform/uniform_{i}.jpg")
  plt.show()
  plt.close()
```

#### Экспоненциальное распределение

```python
for i in range(m):
  # на каком отрезке сторим
  x = np.linspace(0, 20)
  # коэффициент масштаба
  lam = np.random.uniform(low=0, high=5)
  # подсчёт вероятностной функции
  y = expon.pdf(x, scale=lam)
  # отрисовка
  plt.plot(x, y)
  plt.fill_between(x, 0, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./expon/expon_{i}.jpg")
  plt.show()
  plt.close()
```

#### Распределение хи-квадрат

```python
for i in range(m):
  # на каком отрезке строим
  x1 = np.random.randint(1, 100)
  x = np.linspace(0, x1, 1000)
  # кол-во степеней свободы
  df = np.random.randint(1, 5)
  # подсчёт вероятностной функции
  y = chi2.pdf(x, df)
  # отрисовка
  plt.plot(x, y)
  plt.fill_between(x, 0, y)
  # сохранение картинки в нужную папку
  plt.savefig(f"./chi2/chi2_{i}.jpg")
  plt.show()
  plt.close()
```
